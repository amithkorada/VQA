```python
import pandas as pd
from bert_score import score
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('vqa_results_val_10000.csv')
df = df.fillna("")
predicted_answers = df['predicted_answer'].to_list()
ground_truth_answers = df['ground_truth_answer'].to_list()
P, R, F1 = score(predicted_answers, ground_truth_answers, lang="en", model_type="bert-base-uncased")
```

```python
mean_bert_score = np.mean(F1.numpy())
print("Mean BERTScore:", mean_bert_score)
```

```python
accuracy = (df['predicted_answer'] == df['ground_truth_answer']).mean()

# Print the accuracy
print(accuracy)
```
